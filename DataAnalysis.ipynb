```python
# ADF 검정

from statsmodels.tsa.stattools import adfuller
# ADF 테스트 결과 저장할 딕셔너리
adf_results = {}

# 각 컬럼에 대해 ADF 테스트 실행
for column in df.columns:
    if column == 'Date':
        continue  # 날짜 컬럼은 제외
    result = adfuller(df[column].dropna())  # NaN 값 제거 후 ADF 테스트 실행
    adf_results[column] = result
    print(f"ADF Test Results for {column}:")
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    print(f"Critical Values: {result[4]}")
    if result[1] > 0.05:
        print(f"Column {column} has a unit root (p-value > 0.05)")
    else:
        print(f"Column {column} does not have a unit root (p-value <= 0.05)")
    print("\n")

print("ADF 테스트가 완료되었습니다.")
```

```python
# ADF 테스트 결과 저장할 딕셔너리
adf_results = {}

# 각 컬럼에 대해 ADF 테스트 실행 및 차분 후 재검정
for column in df.columns:
    if column == 'Date':
        continue  # 날짜 컬럼은 제외

    # ADF 테스트 실행
    result = adfuller(df[column].dropna())  # NaN 값 제거 후 ADF 테스트 실행
    adf_results[column] = result

    # p-value가 0.05를 넘는 경우 차분
    if result[1] > 0.05:
        # 차분 수행
        df[column] = df[column].diff().dropna()

        # 차분 후 다시 ADF 테스트 실행
        result_diff = adfuller(df[column].dropna())
        adf_results[column] = result_diff

        # 차분된 결과를 원래 DataFrame의 해당 컬럼에 저장
        df[column] = df[column]

        print(f"ADF Test Results after differencing for {column}:")
        print(f"ADF Statistic: {result_diff[0]}")
        print(f"p-value: {result_diff[1]}")
        print(f"Critical Values: {result_diff[4]}")

        if result_diff[1] > 0.05:
            print(f"Column {column} still has a unit root after differencing (p-value > 0.05)")
        else:
            print(f"Column {column} does not have a unit root after differencing (p-value <= 0.05)")
    else:
        print(f"Column {column} does not have a unit root (p-value <= 0.05)")

    print("\n")

print("차분 및 재검정이 완료되었습니다.")
```
